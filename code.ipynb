```python

class Route:
    def __init__(self, route_id, start, end, stops):
        self.route_id = route_id
        self.start = start
        self.end = end
        self.stops = stops

    def display(self):
        print("\n----------------------------")
        print(f"ğŸšŒ Route ID : {self.route_id}")
        print(f"ğŸ“ Start    : {self.start}")
        print(f"ğŸ End      : {self.end}")
        print(f"ğŸ›‘ Stops    : {', '.join(self.stops)}")
        print("----------------------------")


class Bus:
    def __init__(self, bus_id, capacity):
        self.bus_id = bus_id
        self.capacity = capacity
        self.route = None

    def assign_route(self, route):
        self.route = route

    def display(self):
        print("\n----------------------------")
        print(f"ğŸš Bus ID      : {self.bus_id}")
        print(f"ğŸ‘¥ Capacity    : {self.capacity}")
        if self.route:
            print(f"ğŸ›£ Route       : {self.route.route_id} ({self.route.start} â†’ {self.route.end})")
        else:
            print("ğŸ›£ Route       : Not assigned yet")
        print("----------------------------")


class BusRouteManagementSystem:
    def __init__(self):
        self.routes = {}
        self.buses = {}




    def add_route(self):
        print("\nâœ¨ Adding a New Route...")
        route_id = input("Enter a unique Route ID: ")

        if route_id in self.routes:
            print("âš  A route with that ID already exists!")
            return

        start = input("Where does the route start? ")
        end = input("Where does the route end? ")
        stops = input("List the stops (comma separated): ").split(",")
        stops = [s.strip() for s in stops]

        self.routes[route_id] = Route(route_id, start, end, stops)
        print("âœ… Route added successfully!")




    def view_routes(self):
        if not self.routes:
            print("\nğŸ˜• No routes have been added yet.")
            return

        print("\nğŸ—º ALL ROUTES")
        for route in self.routes.values():
            route.display()



    def add_bus(self):
        print("\nâœ¨ Adding a New Bus...")
        bus_id = input("Enter a unique Bus ID: ")

        if bus_id in self.buses:
            print("âš  A bus with that ID already exists!")
            return

        try:
            capacity = int(input("Enter the bus seating capacity: "))
        except ValueError:
            print("âŒ That's not a valid number!")
            return

        self.buses[bus_id] = Bus(bus_id, capacity)
        print("âœ… Bus added successfully!")



    def view_buses(self):
        if not self.buses:
            print("\nğŸ˜• No buses have been added yet.")
            return

        print("\nğŸš ALL BUSES")
        for bus in self.buses.values():
            bus.display()




    def assign_bus_to_route(self):
        print("\nğŸ”— Assigning Bus to Route...")

        bus_id = input("Enter Bus ID: ")
        route_id = input("Enter Route ID: ")

        if bus_id not in self.buses:
            print("âŒ That Bus ID does not exist!")
            return

        if route_id not in self.routes:
            print("âŒ That Route ID does not exist!")
            return

        bus = self.buses[bus_id]
        route = self.routes[route_id]
        bus.assign_route(route)

        print(f"âœ… Bus {bus_id} is now assigned to Route {route_id}!")




    def search_route(self):
        print("\nğŸ” Searching for a Route...")
        start = input("Enter starting location: ")
        end = input("Enter ending location: ")

        found = False
        for route in self.routes.values():
            if route.start.lower() == start.lower() and route.end.lower() == end.lower():
                print("\nğŸ‰ Found a matching route!")
                route.display()
                found = True

        if not found:
            print("\nğŸ˜ No route matches your search.")


    def menu(self):
        while True:
            print("\n====================================")
            print(" ğŸš BUS ROUTE MANAGEMENT SYSTEM")
            print("====================================")
            print("1. Add Route")
            print("2. View Routes")
            print("3. Add Bus")
            print("4. View Buses")
            print("5. Assign Bus to Route")
            print("6. Search Route")
            print("7. Exit")
            print("====================================")

            choice = input("Choose an option (1â€“7): ")

            if choice == "1":
                self.add_route()
            elif choice == "2":
                self.view_routes()
            elif choice == "3":
                self.add_bus()
            elif choice == "4":
                self.view_buses()
            elif choice == "5":
                self.assign_bus_to_route()
            elif choice == "6":
                self.search_route()
            elif choice == "7":
                print("\nğŸ‘‹ Exiting the system. Have a great day!")
                break
            else:
                print("âŒ Invalid choice â€” please pick a number from 1 to 7.")



if __name__ == "__main__":
    system = BusRouteManagementSystem()
    system.menu()
```
